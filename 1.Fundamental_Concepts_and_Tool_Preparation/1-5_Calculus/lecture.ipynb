```python
import numpy as np
import matplotlib.pyplot as plt

# 
plt.rcParams['font.family'] = 'Arial Unicode MS' 

# 
def f(x, y):
    return np.exp(-0.1*(x**2 + y**2)) * (np.sin(x) + np.cos(y))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(14, 6))

# 3D
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.85)
ax1.set_title('3D', fontsize=14)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')

# 
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='viridis', alpha=0.3)
ax2.set_title('', fontsize=14)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x, y)')

#  x y=0
y0 = 0
x_line = x
z_x = f(x_line, y0)
ax2.plot(x_line, np.full_like(x_line, y0), z_x, color='red', lw=3, label='x')
ax2.scatter(x_line, np.full_like(x_line, y0), z_x, color='red', s=10)

#  y x=0
x0 = 0
y_line = y
z_y = f(x0, y_line)
ax2.plot(np.full_like(y_line, x0), y_line, z_y, color='blue', lw=3, label='y')
ax2.scatter(np.full_like(y_line, x0), y_line, z_y, color='blue', s=10)

ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
