```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.family'] = 'Arial Unicode MS' 

# 构造一个“山”型多元函数
def f(x, y):
    return np.exp(-0.1*(x**2 + y**2)) * (np.sin(x) + np.cos(y))

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(14, 6))

# 左图：3D山
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.85)
ax1.set_title('多元函数曲面：一座3D“山”', fontsize=14)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')

# 右图：剖面坡度（偏导数）
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='viridis', alpha=0.3)
ax2.set_title('偏导数：只看单方向的变化', fontsize=14)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_zlabel('f(x, y)')

# 沿 x 方向的剖面（y=0）
y0 = 0
x_line = x
z_x = f(x_line, y0)
ax2.plot(x_line, np.full_like(x_line, y0), z_x, color='red', lw=3, label='x方向剖面')
ax2.scatter(x_line, np.full_like(x_line, y0), z_x, color='red', s=10)

# 沿 y 方向的剖面（x=0）
x0 = 0
y_line = y
z_y = f(x0, y_line)
ax2.plot(np.full_like(y_line, x0), y_line, z_y, color='blue', lw=3, label='y方向剖面')
ax2.scatter(np.full_like(y_line, x0), y_line, z_y, color='blue', s=10)

ax2.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
